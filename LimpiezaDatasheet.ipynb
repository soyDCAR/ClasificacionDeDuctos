# Remplazar doscs de imagen a img

```python
import os
carpeta = 'img' 

if not os.path.isdir(carpeta):
    print(f"La carpeta '{carpeta}' no existe.")
else:
    archivos_encontrados = os.listdir(carpeta)
    print(f"Archivos encontrados en '{carpeta}': {archivos_encontrados}")
    
    renombrados = 0
    for nombre_archivo in archivos_encontrados:
        if nombre_archivo.lower().endswith('.png') and nombre_archivo.startswith('imagen'):
            nuevo_nombre = nombre_archivo.replace('imagen', 'img', 1)
            ruta_original = os.path.join(carpeta, nombre_archivo)
            ruta_nueva = os.path.join(carpeta, nuevo_nombre)

            os.rename(ruta_original, ruta_nueva)
            print(f'Renombrado: {nombre_archivo} -> {nuevo_nombre}')
            renombrados += 1

    if renombrados == 0:
        print("No se encontrÃ³ ningÃºn archivo .png que empiece con 'imagen'.")
    else:
        print(f"âœ… Total de archivos renombrados: {renombrados}")
```

# Conteo de archivos

```python
import os
import re

carpeta = 'img'

patron = re.compile(r'^img(\d+)_d\d+_o\d+_v\d+\.(png|jpg|jpeg)$', re.IGNORECASE)

numeros_img = []

for archivo in os.listdir(carpeta):
    match = patron.match(archivo)
    if match:
        numero = int(match.group(1))
        numeros_img.append(numero)

# Set con todos los nÃºmeros esperados
esperados = set(range(490))
presentes = set(numeros_img)

faltantes = sorted(esperados - presentes)
duplicados = [num for num in numeros_img if numeros_img.count(num) > 1]

print(f'ğŸ”¢ Total Ãºnicos encontrados: {len(presentes)}')
print(f'ğŸ“› NÃºmeros faltantes en la secuencia (0â€“489): {faltantes}')

if duplicados:
    print(f'âš ï¸ NÃºmeros duplicados: {sorted(set(duplicados))}')
```

# Cambiar ruta a minusculas

```python
import os

carpeta = 'img'  # Cambia la ruta si es necesario

# Extensiones que queremos convertir a minÃºscula
extensiones = ['.PNG', '.JPG', '.JPEG']

# Recorremos los archivos en la carpeta
for archivo in os.listdir(carpeta):
    nombre_base, extension = os.path.splitext(archivo)
    if extension.upper() in extensiones:
        nueva_extension = extension.lower()
        nuevo_nombre = nombre_base + nueva_extension

        ruta_original = os.path.join(carpeta, archivo)
        ruta_nueva = os.path.join(carpeta, nuevo_nombre)

        os.rename(ruta_original, ruta_nueva)
        print(f'âœ… Renombrado: {archivo} -> {nuevo_nombre}')
```
